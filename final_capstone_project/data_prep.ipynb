```python
temp1 = df[df['title'].str.find('Men') >= 0]
temp2 = df[df['title'].str.find('Women') >= 0]
temp3 = tf[tf['title'].str.find('Men') >= 0]
temp4 = tf[tf['title'].str.find('Women') >= 0]
```

```python
for x in temp1.index:
    df.at[x, 'maincateg'] = "Men"

for x in temp2.index:
    df.at[x, 'maincateg'] = "Women"

for x in temp3.index:
    tf.at[x, 'maincateg'] = "Men"

for x in temp4.index:
    tf.at[x, 'maincateg'] = "Women"
```

```python
df['maincateg'].isna().sum()
```

```python
tf['maincateg'].isna().sum()
```

```python
df['maincateg'] = df['maincateg'].fillna('Women')
tf['maincateg'] = tf['maincateg'].fillna('Women')
```

```python
df.isna().sum()
```

```python
tf.isna().sum()
```

```python
df.drop(['id', 'title', 'Offer %'], axis=1, inplace=True)
tf.drop(['id', 'title'], axis=1, inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['maincateg'] = le.fit_transform(df['maincateg'])
df['platform'] = le.fit_transform(df['platform'])
tf['maincateg'] = le.fit_transform(tf['maincateg'])
tf['platform'] = le.fit_transform(tf['platform'])
```
