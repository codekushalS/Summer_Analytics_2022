```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, plot_confusion_matrix, ConfusionMatrixDisplay, accuracy_score, roc_auc_score
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
pd.options.mode.chained_assignment = None
%matplotlib inline
```

```python
train_data = pd.read_csv("Train_data.csv")
train_data["GP"] = train_data["GP"].astype(int)
train_data.head()
```

```python
test_data = pd.read_csv("Test_data.csv")
test_data["GP"] = test_data["GP"].astype(int)
test_data.head()
test_data.shape
```

```python
X_train = train_data.drop(["Target", "FG%", "3P%", "FT%", "REB"], axis=1)
y_train = train_data["Target"]
X_test = test_data.drop(["FG%", "3P%", "FT%", "REB"], axis=1)
```

```python
X_train.head()
```

```python
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor

model1 = DecisionTreeRegressor()
model1.fit(X_train, y_train)
y1pred = model1.predict(X_test)
y1pred[y1pred>0.5] = 1
y1pred[y1pred<=0.5] = 0
y1pred = y1pred.astype(int)
y1pred
```

```python
ypred = clf.predict(X_test)
```

```python
#decision tree regressor 98.916
#scaled decision tree 98.916
#
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.tree import DecisionTreeClassifier

model2 = make_pipeline(StandardScaler(), DecisionTreeRegressor())
model2.fit(X_train, y_train)
y2pred = model2.predict(X_test)
y2pred[y2pred>0.5] =1
y2pred[y2pred<=0.5] = 0
y2pred = y2pred.astype(int)
```

```python
res = pd.DataFrame(y2pred)
res.index = X_test.index
res.columns = ["prediction"]
res.to_csv("dectreg3.csv", index=False)
```
