```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, plot_confusion_matrix, classification_report, ConfusionMatrixDisplay, accuracy_score, roc_auc_score
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_csv("Train_data.csv")
df["GP"] = df["GP"].astype(int)
df.head()
```

```python
X = df.drop("Target", axis=1)
y=df["Target"]
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
column_scaled = ["GP", "MIN", "PTS", "FGA", "FG%", "3P%", "FT%"]

X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

for col in column_scaled:
    scale = StandardScaler().fit(X_train[[col]])
    scale = StandardScaler().fit(X_test[[col]])
    X_train_scaled[col] = scale.transform(X_train[[col]])
    X_test_scaled[col] = scale.transform(X_test[[col]])
```

```python
tree= DecisionTreeClassifier(random_state=2) # using the random state for reproducibility
forest= RandomForestClassifier(random_state=2)
knn= KNeighborsClassifier()
svm= SVC(random_state=2)
xboost= XGBClassifier(random_state=2)

# now, create a list with the objects 
models= [tree, forest, knn, svm, xboost]

for model in models:
    model.fit(X_train, y_train) # fit the model
    y_pred= model.predict(X_test) # then predict on the test set
    accuracy= accuracy_score(y_test, y_pred) # this gives us how often the algorithm predicted correctly
    # clf_report= classification_report(y_test, y_pred) # with the report, we have a bigger picture, with precision and recall for each class
    print(f"The accuracy of model {type(model).__name__} is {accuracy:.5f}")
    # print(clf_report)
    print("\n")
```
